```python
import pandas as pd
import numpy as np
import tqdm
```

```python
best_without_outlier_sub = pd.read_csv("去除异常值_sub/submission_1.5540623551602637.csv")
```

```python
best_classifer_model_sub = pd.read_csv("class_sub/submission_0.9038386857136778.csv")
```

```python
best_classifer_model_sub.sort_values(by='target',ascending = False,inplace=True)
```

```python
best_classifer_model_sub.head(20)
```

```python
123623*0.0106
```

```python
#In case missing some predictable outlier, we choose top 25000 with highest outliers likelyhood.
outlier_id = pd.DataFrame(best_classifer_model_sub.sort_values(by='target',ascending = False).head(25000)['card_id'])
```

```python
outlier_id
```

```python
best_submission = pd.read_csv('./sub/2_blend.csv')
```

```python
most_likely_liers = best_submission.merge(outlier_id,how='right')
most_likely_liers.head()
```

```python
eda_most_like =  most_likely_liers.copy()
tmp = best_classifer_model_sub.copy()
tmp.rename(columns={'target':'prob'},inplace=True)
eda_most_like = pd.merge(eda_most_like,tmp,how='left',on='card_id')
```

```python
eda_most_like.sort_values(by='prob',ascending = False,inplace=True)
eda_most_like.head(50)
```

```python
eda_most_like['count_06'] = eda_most_like['prob'].map(lambda x : 1 if x>0.4  else  0)
```

```python
eda_most_like['count_06'].value_counts()
```

```python
eda_most_like['target'] = eda_most_like.apply(lambda x:  -33.21928095  if x.prob>=0.4 else x.target,axis=1)
```

```python
eda_most_like.head(70)
```

```python
most_likely_liers =  eda_most_like[['card_id','target']]
```

```python
most_likely_liers.head(60)
```

```python
for card_id in tqdm.tqdm(most_likely_liers['card_id']):
    best_without_outlier_sub.loc[best_without_outlier_sub['card_id']==card_id,'target']\
    = most_likely_liers.loc[most_likely_liers['card_id']==card_id,'target'].values
```

```python
best_without_outlier_sub.to_csv("combining_submission.csv", index=False)
```
